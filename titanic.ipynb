What was the gender-based disparity in survival rates among titanic passengers?

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the Titanic dataset from a CSV file
df = pd.read_csv('titanic.csv')

# Creating separate DataFrames for male and female passengers
male_df = df[df['Sex'] == 'male']
female_df = df[df['Sex'] == 'female']

# Calculating the survival counts for male and female passengers
survivalcount_male = male_df['Survived'].value_counts()
survivalcount_female = female_df['Survived'].value_counts()

# Combine the survival counts for both genders into a single list
combined_counts = [survivalcount_male[0], survivalcount_male[1], survivalcount_female[0], survivalcount_female[1]]

# Define labels for the pie chart segments
labels = ['Male Died', 'Male Lived', 'Female Died', 'Female Lived']

# Making an 'explode' parameter to emphasize a portion of the pie chart
explode = (0, 0.1, 0, 0.1)

# Create a pie chart with the combined survival counts
plt.pie(combined_counts, labels=labels, explode=explode, autopct='%1.1f%%', startangle=140)

# Set the aspect ratio of the pie chart to be equal to make it a circle
plt.axis('equal')

# Set a title for the pie chart
plt.title('Survival Percentage Among Genders')

#  pie chart
plt.show()
```

What was the survival rate of those who survived under 18?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Titanic dataset from a CSV file
df = pd.read_csv('titanic.csv')

# Filter the DataFrame to select passengers under the age of 18
under18 = df[df['Age'] < 18]

# Calculate the survival rate for passengers under 18
survival_rate = (under18['Survived'] == 1).mean()

# Create a bar chart to visualize the survival rate
plt.bar(['Under 18'], [survival_rate], color=['blue'])
plt.ylabel('Survival Rate')
plt.title('Survival Rate for Passengers Under 18')
plt.ylim(0, 1)  # Set the y-axis limits to 0 and 1

# Display the survival rate as text on the chart
plt.text('Under 18', survival_rate, f'{survival_rate:.2%}', ha='center', va='bottom')

# Show the bar chart
plt.show()
```
