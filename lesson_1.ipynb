```python
# Import the matplotlib library for data visualization.
import matplotlib.pyplot as plt

# Define a dictionary named GoodEats with food items as keys and the amount consumed as values.
GoodEats = {'Bacon': 50, 'Sandwich': 10, 'Eggs': 40, 'Chicken': 5}

# Create a list of food items (keys) from the GoodEats dictionary.
food = list(GoodEats.keys())

# Create a list of the corresponding amounts consumed aka Values from the GoodEats dictionary.
values = list(GoodEats.values())

# Create a figure for the plot with a specified size (width and height).
fig = plt.figure(figsize=(10, 5))

# Create a bar chart using the data in the 'food' and 'values' lists. Set the bar color to 'maroon' and the width to 0.4.
plt.bar(food, values, color='maroon', width=0.4)

# Add a label to the x-axis of the chart.
plt.xlabel("Food")

# Add a label to the y-axis of the chart.
plt.ylabel("Amount Ate")

# Add a title to the chart.
plt.title("Most Popular in School")

# Display the chart.
plt.show()

```
