Using your knowledge of Python and Matplotlib write your code to transform and graph the data below.

After your code create a new Markdown block to describe your analysis

````python
import matplotlib.pyplot as plt

# Initialize counts for each state
NY = 0
FL = 0
MO = 0
TX = 0
KS = 0

# Open the file 'states.txt'
with open('states.txt', 'r') as f:
    for row in f:
        # Remove leading and trailing whitespace
        row = row.strip()
        if row == "NY":
            NY += 1
        elif row == "MO":
            MO += 1
        elif row == 'FL':
            FL += 1
        elif row == 'TX':
            TX += 1
        elif row == 'KS':
            KS += 1

# Create lists for state names and their respective counts
state_names = ['NY', 'FL', 'MO', 'TX', 'KS']
state_counts = [NY, FL, MO, TX, KS]

# Create a bar chart
plt.bar(state_names, state_counts, label='Data of States')

# Label the axes and add a title
plt.xlabel('Name of States', fontsize=11)
plt.ylabel('Amount', fontsize=11)
plt.title('Occurrences of State Abbreviations in File', fontsize=11)
plt.legend()

# Show the chart
plt.show()

````
